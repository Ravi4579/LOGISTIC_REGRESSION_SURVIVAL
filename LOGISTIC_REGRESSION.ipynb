```python
#Reading dataset
#dft=train dataset
dft=pd.read_csv('Titanic_train.csv')
#dfp=test dataset(prediction dataset)
dfp=pd.read_csv('Titanic_test.csv')
```

```python
#filling null values
#for train set
dft['Age'].fillna(dft['Age'].median(),inplace=True)             #numerical column
dft['Embarked'].fillna(dft['Embarked'].mode()[0],inplace=True)  #categorical column
```

```python
# Create a binary feature indicating the presence of Cabin information
#for train dataset
dft['Cabin_Missing'] = dft['Cabin'].notnull().astype(int)
#for test dataset
dfp['Cabin_Missing'] = dfp['Cabin'].notnull().astype(int)
```

```python
# Apply the condition to create the new column 'Age'
#for train dataset
dft['Sex'] = np.where(dft['Sex'] == 'male', 0, 1)
#for test dataset
dfp['Sex'] = np.where(dfp['Sex'] == 'male', 0, 1)
```

```python
#Checking for duplicates
dft.duplicated().sum()
```

```python
#creating pair plot
import matplotlib.pyplot as plt
sns.pairplot(dft)
plt.show()
```

```python
dft.info()
```

```python
#non useful data
#for train dataset
dft1=dft.drop(['PassengerId','Name','Ticket','Cabin'],axis=1)
#for test dataset
dfp1=dfp.drop(['PassengerId','Name','Ticket','Cabin'],axis=1)
```

```python
# Rename the 'Cabin_Missing' column to 'cabin'
dft1.rename(columns={'Cabin_Missing': 'cabin'}, inplace=True)
dfp1.rename(columns={'Cabin_Missing': 'cabin'}, inplace=True)
```

```python
dft1.info()
```

```python
# Creating dummy variables for the 'Embarked' column
#for train set
dft1 = pd.get_dummies(dft1, columns=['Embarked'], prefix='Embarked', drop_first=True)
#for test set
dfp1 = pd.get_dummies(dfp1, columns=['Embarked'], prefix='Embarked', drop_first=True)
```

```python
# Convert any boolean columns to integers (0 and 1)
#for train set
dft1[['Embarked_Q','Embarked_S']] = dft1[['Embarked_Q','Embarked_S']].applymap(lambda x: 1 if x is True else 0)
#for test set
dfp1[['Embarked_Q','Embarked_S']] = dfp1[['Embarked_Q','Embarked_S']].applymap(lambda x: 1 if x is True else 0)
```

```python
#deleting outliers
# Define a function to remove outliers based on IQR
def remove_outliers_iqr(dft1, column):
    Q1 = dft1[column].quantile(0.25)
    Q3 = dft1[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df_filtered = dft1[(dft1[column] >= lower_bound) & (dft1[column] <= upper_bound)]
    return df_filtered

# Remove outliers from 'Age' and 'Fare' columns in dft1
dft1 = remove_outliers_iqr(dft1, 'Age')
dft1 = remove_outliers_iqr(dft1, 'Fare')
```

```python
# preprocessing and standardization

from sklearn.preprocessing import StandardScaler
import pickle

# Assuming 'dft1' is your DataFrame and you want to standardize numerical features
numerical_cols = ['Pclass', 'Sex', 'Age', 'SibSp','Parch', 'Fare', 'cabin','Embarked_Q', 'Embarked_S']  # Replace with your actual numerical columns

# Create a StandardScaler object
scaler = StandardScaler()

# Fit and transform the numerical features
#for train set
dft1[numerical_cols] = scaler.fit_transform(dft1[numerical_cols])
#for test set
dfp1[numerical_cols] = scaler.transform(dfp1[numerical_cols])

# Now dft1 has standardized numerical features
```

# Model building

```python
#importing library
from sklearn.linear_model import LogisticRegression
#specifying dependent and independent variables
X=dft1.iloc[:,1:10]  #independent
Y=dft1.iloc[:,0]     #dependent
classifier = LogisticRegression()
classifier.fit(X,Y)
```

```python
# we want to find the best hyperparameters for 'C' and 'penalty'
from sklearn.model_selection import GridSearchCV

# Define the parameter grid to search
param_grid = {
    'C': [0.1, 1, 10,20,50],  # Regularization strength
    'penalty': ['l1', 'l2'],  # Regularization type
    'max_iter': [100,200,300]
}

# Create a GridSearchCV object
classifier_reg = GridSearchCV(classifier, param_grid=param_grid, cv=5)  # 5-fold cross-validation

# Fit the GridSearchCV object to your data
classifier_reg.fit(X, Y)

# Access the best hyperparameters
best_params = classifier_reg.best_params_
print(f"Best Hyperparameters: {best_params}")

# Now you can use the best hyperparameters to create your final model
best_classifier = LogisticRegression(**best_params)
best_classifier.fit(X, Y)
```

```python
#predicting train set
y_pred=classifier_reg.predict(X)
y_pred
```

```python
#model after performing hyperparameter tuning or cross-validation.
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
classifier_reg.best_score_
```

```python
# Accuracy of a mode
score=accuracy_score(Y,y_pred)
score
```

```python
print(classification_report(Y,y_pred))
```

```python
# create ROC-AUC score.
from sklearn.metrics import roc_auc_score

# Calculate the ROC AUC score
roc_auc = roc_auc_score(Y, y_pred) # Use y_pred_proba if necessary

print(f"ROC AUC Score: {roc_auc}")
```

```python
# viz roc curve
from sklearn.metrics import roc_curve, auc

fpr, tpr, thresholds = roc_curve(Y, y_pred)
roc_auc = auc(fpr, tpr)

plt.figure()
plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC)')
plt.legend(loc="lower right")
plt.show()
```
